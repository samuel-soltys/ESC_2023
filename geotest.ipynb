```python
import geopandas as gpd
import geoplot
import geoplot.crs as gcrs
import matplotlib.pyplot as plt
```

```python
map = gpd.read_file("okresy.json")
geoplot.polyplot(map, projection=gcrs.AlbersEqualArea())
```

```python
map.shape
```

```python
import numpy as np
map['heatmap'] = np.random.randint(1, 60, map.shape[0])
```

```python
import mapclassify
heatmap_data = map["heatmap"]
scheme = mapclassify.Quantiles(heatmap_data, k=20)

geoplot.choropleth(
    map, hue=heatmap_data, scheme=scheme, cmap='Blues', figsize=(10, 12), projection=gcrs.AlbersEqualArea()
)
```
